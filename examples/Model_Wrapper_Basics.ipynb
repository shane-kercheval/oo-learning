```python
import pandas as pd
data = pd.read_csv('./data/cement.csv')
data_x = data.drop('strength', axis=1)
data_y = data.strength
```

```python
# !pip install oolearning --upgrade
```

```python
from oolearning import *
model = RegressionMW()
model.train(data_x=data_x, data_y=data_y)
```

```python
model.fitted_info.results_summary
```

```python
model.fitted_info.summary_stats
```

```python
predictions = model.predict(data_x=data_x)
RmseEvaluator().evaluate(actual_values=data_y, predicted_values=predictions)
```
